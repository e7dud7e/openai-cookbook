## Semantic text search using embeddings

We can search through all our reviews semantically in a very efficient manner and at very low cost, by embedding our search query, and then finding the most similar reviews. The dataset is created in the [Obtain_dataset Notebook](Obtain_dataset.ipynb).

```python
import pandas as pd
import numpy as np

datafile_path = "data/fine_food_reviews_with_embeddings_1k.csv"

df = pd.read_csv(datafile_path)

df
```

```python
df["embedding"] = df.embedding.apply(eval).apply(np.array)
df
```

Here we compare the cosine similarity of the embeddings of the query and the documents, and show top_n best matches.

```python
from openai.embeddings_utils import get_embedding, cosine_similarity
```

```python
txt1 = "I am happy"
txt2 = "We are glad"
txt3 = "life is like a box of chocolates"
emb1 = get_embedding(txt1, engine="text-embedding-ada-002") # these return a list of 1536 numbers
emb2 = get_embedding(txt2, engine="text-embedding-ada-002")
emb3 = get_embedding(txt3, engine="text-embedding-ada-002")
```

```python
cosine_similarity(emb1,emb2)
```

```python
cosine_similarity(emb1,emb3)
```

```python
cosine_similarity(emb2,emb3)
```

```python
product_description = "good soup dumplings"
product_embedding = get_embedding(product_description, engine="text-embedding-ada-002")
```

#### Make a new column "similarity"
Reuse this column for every search by computing the cosine similarity of the product description with the embedding of maybe "combined", which is title and summary combined.

```python
df["similarity"] = df['embedding'].apply(lambda emb: cosine_similarity(emb, product_embedding))
df
```

#### Sort and get the top n results

```python
# set dataframe display width
pd.set_option('display.max_colwidth', 200)
df_sorted = df.sort_values('similarity', ascending=False)
df_sorted
```

```python
# set dataframe display width
pd.set_option('display.max_colwidth', 500)
df_sorted[['combined','similarity']]
```

```python
# show all text (avoid truncation)
pd.set_option('display.max_colwidth', None)
df_sorted[['combined','similarity']]
```

```python
# search through the reviews for a specific product
def search_reviews(df, product_description, n=3, pprint=True):
    product_embedding = get_embedding(
        product_description,
        engine="text-embedding-ada-002"
    )
    df["similarity"] = df.embedding.apply(lambda x: cosine_similarity(x, product_embedding))

    results = (
        df.sort_values("similarity", ascending=False)
        .head(n)
        .combined.str.replace("Title: ", "")
        .str.replace("; Content:", ": ")
    )
    if pprint:
        for r in results:
            print(r[:200])
            print()
    return results


# results = search_reviews(df, "delicious beans", n=3)
results = search_reviews(df, "good soup dumplings", n=3)
```

```python
results = search_reviews(df, "whole wheat pasta", n=3)
```

We can search through these reviews easily. To speed up computation, we can use a special algorithm, aimed at faster search through embeddings.

```python
results = search_reviews(df, "bad delivery", n=1)
```

As we can see, this can immediately deliver a lot of value. In this example we show being able to quickly find the examples of delivery failures.

```python
results = search_reviews(df, "spoilt", n=1)
```
